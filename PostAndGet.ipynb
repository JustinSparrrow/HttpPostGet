```python
# 3. post请求举例 -- 金山翻译

# 示例翻译：学习 -- study
# url为： https://ifanyi.iciba.com/index.php?c=trans&m=fy&client=6&auth_user=key_web_fanyi&sign=85e608fb1c4bbdd4
# 直接在网页中访问url会显示 {"error_code":10005,"message":"auth failed"} ，这是因为该访问为post需要携带参数
# post携带from data的内容放进参数字典里面

url = 'https://ifanyi.iciba.com/index.php?c=trans'

# 构建请求头
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'
}

# q表示传进去的汉语，可以将其变为变量，实现输入
word = input('请输入你要翻译的内容')
# 构建data参数字典
post_data = {
    'from': 'zh',
    'to': 'en',
    'q': word
}
# 发送请求
res = requests.post(url, headers = headers, data = post_data)
print(res.text)

# 解析数据
# 将json数据转换为python字典
import json

# json格式转换为python格式
dic = json.loads(res.text)
# 检测格式
print(type(dic))  # 检测之后是字典格式
print(dic['out'])  # 通过输出字典的方式输出值，json中与翻译内容对应的键为'out'
```
