## 1. 获取网页内容  
http:超文本传输协议，客户端与服务端之间的请求-响应协议  
最常见的是GET方法和POST方法

一个完整的HTTP请求如下：  
```markdown
**请求行：**  
Post /user/info HTTP/1.1  

**请求头：**  
Host: www.example.com  
User-Agent: curl/7.77.0  
Accept: */ *  

**请求体（POST中传递数据）：**  
{"username": "moqi",  
"email": "moqi@hotmail.com"}  
```

http响应：  
```markdown
状态行:  
HTTP/1.1 200 OK  

响应头：  
Date: Fri, 27 Jan 2023 02:10:48 GMT  
Content-Type:text/html; charset=utf-8  

响应体：  
<!DOCTYPE html>  
    <head><title>首页</title></head>  
    <body><h1>moqi</h1><p>hello!</p></body?
<html>
```

````python
import requests
````

````python
response_book = requests.get("http://books.toscrape.com/")
# print(response.status_code)     # 200说明请求成功
if response_book.ok:
    print(response_book.text)
else:
    print("请求失败")
````

````python
# 使用header将请求伪装成浏览器请求
head = {"User-Agent": "Mozilla/5.0 (Windows NT 11.0; Win64; x64"}
response_book = requests.get("http://books.toscrape.com/", headers=head)
if response_book.ok:
    print(response_book.text)
else:
    print("Error!")
````

````python
head = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36 SLBrowser/9.0.0.10191 SLBChan/105"}
response_movie = requests.get("http://movie.douban.com/top250", headers=head)
print(response_movie.status_code)
print(response_movie.text)
````

![一般情况解析后的HTML树](/reptile/assets/img.png)

````python
content_book = requests.get("http://books.toscrape.com/").text
soup_book = BeautifulSoup(content_book, "html.parser")
all_prices = soup_book.find_all("p", attrs={"class": "price_color"})
all_titles = soup_book.find_all("h3")
for title, price in zip(all_titles, all_prices):
    link = title.find("a")["href"]  # 获取书籍链接
    title_text = title.find("a").text  # 获取书籍标题
    price_text = price.string  # 获取价格文本
    print(f"Title: {title_text:<50} Price: {price_text:<10} Link: {link}")     
````

````python
html = response_movie.text
soup_movie = BeautifulSoup(html, "html.parser")
all_titles = soup_movie.find_all("span", attrs={"class": "title"})
for title in all_titles:
    # print(title.string)
    title_string = title.string
    if "/" not in title_string:
        print(title_string)
````

````python
i = 1
for start_num in range(0, 250, 25):
    response_movie = requests.get(f"http://movie.douban.com/top250?start={start_num}", headers=head)
    html = response_movie.text
    soup_movie = BeautifulSoup(html, "html.parser")
    all_titles = soup_movie.find_all("span", attrs={"class": "title"})
    for title in all_titles:
        # print(title.string)
        title_string = title.string
        if "/" not in title_string:
            print(f"{i}.{title_string}")
            i += 1
````
